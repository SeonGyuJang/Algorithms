```python
def multiply_bin(x_bin: str, y_bin: str, depth: int = 0, tag: str = "") -> int:
    indent = "  " * depth
    if tag:
        print(f"{indent}[{tag}]")

    n = max(len(x_bin), len(y_bin))
    if n == 1:
        result = int(x_bin, 2) * int(y_bin, 2)
        print(f"{indent}{x_bin} × {y_bin} = {result:b}_2")
        return result

    x_bin = x_bin.zfill(n)
    y_bin = y_bin.zfill(n)

    m = (n + 1) // 2
    x_L, x_R = x_bin[:-m], x_bin[-m:]
    y_L, y_R = y_bin[:-m], y_bin[-m:]

    print(f"{indent}n={n}, m={m}")
    print(f"{indent}x = {x_bin[:n-m]} {x_bin[n-m:]}  →  x_L={x_L}, x_R={x_R}")
    print(f"{indent}y = {y_bin[:n-m]} {y_bin[n-m:]}  →  y_L={y_L}, y_R={y_R}")

    P1 = multiply_bin(x_L, y_L, depth + 1, "P1")
    P2 = multiply_bin(x_R, y_R, depth + 1, "P2")
    P3 = multiply_bin(
        bin(int(x_L, 2) + int(x_R, 2))[2:],
        bin(int(y_L, 2) + int(y_R, 2))[2:],
        depth + 1,
        "P3",
    )

    result = (P1 << (2 * m)) + ((P3 - P1 - P2) << m) + P2
    print(f"{indent}P1={P1:b}_2, P2={P2:b}_2, P3={P3:b}_2  →  result={result:b}_2")
    return result


if __name__ == "__main__":
    X = "10110110"
    Y = "11001001"
    product = multiply_bin(X, Y)
    print("\n최종 곱:", f"{product:b}_2 (={product})")
    assert product == int(X, 2) * int(Y, 2)
```
